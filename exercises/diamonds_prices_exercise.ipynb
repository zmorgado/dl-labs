```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LinearRegression
from sklearn.metrics import confusion_matrix
from sklearn.neighbors import KNeighborsClassifier # El KNN
from sklearn.preprocessing import StandardScaler, MinMaxScaler

from sklearn.metrics import r2_score
```

```python
# Nos traemos la base de datos:
df = pd.read_csv('../datasets/diamonds_prices_2022.csv')
df.head()
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
X = df.drop('price', axis=1)
y = df['price']
```

```python
X.head()
```

```python
y.head()
```

```python
encoder = LabelEncoder()

X_encoded = X.copy()
```

```python

for column in X_encoded[ ['cut', 'color', 'clarity'] ]:
    print(f'Columna {column}, variables: {list(X_encoded[column].unique())}')
```

```python
for column in X[ ['cut', 'color', 'clarity'] ]:
    X_encoded[column] = encoder.fit_transform(X_encoded[column])
```

```python
for column in X_encoded[ ['cut', 'color', 'clarity'] ]:
    print(f'Columna {column}, variables: {X_encoded[column].unique()}')
```

```python
X_encoded.describe().loc[['mean','std','min','max']]
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.25, random_state=42
)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)  
X_test_scaled = scaler.transform(X_test)
```

```python
arbol_decision = DecisionTreeClassifier(max_depth=7)
```

```python
arbol_decision.fit(X_train_scaled, y_train)
```

```python
arbol_decision.score(X_test_scaled, y_test)
```

```python
model_lr = LinearRegression()
model_lr.fit(X_train_scaled, y_train)
```

```python
# Predicciones
y_predlr = model_lr.predict(X_test_scaled)
```

```python
y_predlr
```

```python
r2 = r2_score(y_test, y_predlr)
r2
```
