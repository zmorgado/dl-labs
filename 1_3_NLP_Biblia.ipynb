```python
import pandas as pd
import nltk
import matplotlib.pyplot as plt

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer
from nltk.probability import FreqDist
from nltk import pos_tag
from nltk.metrics.distance import edit_distance
from nltk.stem import WordNetLemmatizer
from wordcloud import WordCloud

nltk.download('wordnet')
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('cess_esp')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
with open("/content/drive/MyDrive/biblia.txt", 'r', encoding='utf-8') as myfile:
    biblia = myfile.read().replace('\n', ' ')
```

```python
len(biblia)
```

```python
tokens = word_tokenize(biblia, language="spanish")

# Los diez primeros tokens, i.e. pseudo-palabras
tokens[:10]
```

```python
len(tokens)
```

```python
# diez palabras más frecuentes
freq_w_biblia = FreqDist(tokens)
freq_w_biblia.most_common(10)
```

```python
stop_words = set(stopwords.words('spanish'))
```

```python
filtered_tokens = [word.lower() for word in tokens if word.isalnum() and word.lower() not in stop_words]
```

```python
len(set(filtered_tokens))
```

```python
freq_w_filtered_biblia = FreqDist(filtered_tokens)
freq_w_filtered_biblia.most_common(10)
```

#### Stemming

```python
stemmer = SnowballStemmer('spanish')
stemmed_tokens = [stemmer.stem(word) for word in filtered_tokens]
```

```python
stemmed_tokens[123000:123010]
```

```python
freq_w_stemmed_biblia = FreqDist(stemmed_tokens)
freq_w_stemmed_biblia.most_common(10)
```

```python
stemmed_tokens_size = len(stemmed_tokens)
stemmed_tokens_size
```
