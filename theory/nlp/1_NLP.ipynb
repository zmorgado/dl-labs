**¿QUÉ DEBES HABER ENTENDIDO?**

La **tokenización** es el primer paso en el preprocesamiento de textos en NLP. Consiste en dividir un texto continuo en unidades más pequeñas, llamadas **tokens**.

En **procesamiento de lenguaje natural (NLP)**, un **token** es una unidad mínima de análisis en la que dividimos un texto. Normalmente un token corresponde a una palabra, pero no siempre.
* Un token puede ser también:

  * Un **signo de puntuación**: `"."`, `","`
  * Un **número**: `"9.8"`, `"2025"`
  * Una **abreviatura o acrónimo**: `"EE.UU."`, `"ONU"`
  * Una **parte de una palabra** en algunos modelos (por ejemplo, en tokenizadores subpalabra como BPE o WordPiece, *“jugando”* → *\["juga", "ndo"]*).

Un token no tiene que coincidir con una palabra completa porque depende  de la tarea y del método empleado para la tokenización:

1. **Tokenización clásica (palabras)**
   * En NLTK, `word_tokenize("Los estudiantes leen.")` → `["Los", "estudiantes", "leen", "."]`
   * Aquí un token suele coincidir con una palabra o signo de puntuación.

2. **Tokenización subpalabra (ej. BERT, GPT)**
   * Se buscan **fragmentos frecuentes** para reducir vocabulario y manejar palabras desconocidas.
   * Ejemplo: `"increíblemente"` → `["incre", "íble", "mente"]`.
   * Así el modelo puede **entender partes comunes** y **generalizar mejor** a palabras nuevas.

3. **Tokenización a nivel de carácter**
   * Cada letra es un token: `"hola"` → `["h", "o", "l", "a"]`.
   * Útil para lenguajes con muchas variaciones o escritura sin espacios (como chino o japonés).

Por tanto, **token ≠ palabra**, sino que se trata la *unidad mínima de análisis de textos con NLP, y que resulta ser dependiente de la técnica de tokenización elegida*.

Casi todos los algoritmos de NLP (frecuencias, embeddings, modelos de lenguaje, etc.) trabajan con tokens y no con texto en bruto. Ejemplo:
  ```
  Texto original: "Los estudiantes leen libros."
  Tokens: ["Los", "estudiantes", "leen", "libros", "."]
  ```

En este ejercicio utilizaremos la librería **NLTK (Natural Language Toolkit)** de Python. Su función `word_tokenize` implementa un *tokenizador basado en reglas y modelos estadísticos*, que permite segmentar el texto en palabras y signos de puntuación de forma automática.

**¿QUÉ DEBES HABER ENTENDIDO?**

Después de la **tokenización**, los textos suelen contener variaciones y elementos que dificultan el análisis automático. La **normalización** busca **homogeneizar** los tokens para reducir ruido y mejorar la representación del lenguaje.

Las técnicas más habituales son:

1. **Conversión a minúsculas (*lowercasing*)**

   * Evita que “UNED” y “uned” se traten como palabras distintas.
   * Útil en tareas semánticas donde las mayúsculas no aportan información relevante.

2. **Eliminación de caracteres no alfanuméricos**

   * Filtra puntuación, símbolos o caracteres especiales.
   * Se conservan solo letras y números cuando interesa trabajar con palabras limpias.

3. **Eliminación de *stopwords***

   * Son palabras muy frecuentes (ej. “de”, “los”, “en”) que tienen poco valor semántico en ciertos análisis.
   * Al eliminarlas, reducimos el tamaño del vocabulario y nos quedamos con términos más informativos.

Sin embargo, estas decisiones dependen de la tarea:

* En un análisis de estilo o sintaxis, puede ser importante conservar mayúsculas o stopwords.
* En una tarea de clasificación de temas, conviene simplificar para destacar los términos relevantes.

**¿QUÉ DEBES HABER ENTENDIDO?**

Cuando analizamos texto, muchas palabras aparecen en **formas flexionadas** (tiempos verbales, plurales, géneros…). Para simplificar, es útil reducirlas a una forma más general que represente el mismo concepto.

* **Stemming**

  * Consiste en recortar las palabras hasta su **raíz morfológica** mediante reglas simples.
  * Ejemplo: *“correr”, “corriendo”, “corrían” → “corr”*.
  * Es rápido y eficiente, pero puede generar formas que **no son palabras reales**.

* **Lematización**

  * Devuelve la **forma canónica o diccionario** (*lema*) de la palabra, considerando su **contexto gramatical**.
  * Ejemplo: *“corren”, “corría”, “correrán” → “correr”*.
  * Es más precisa y conserva el significado, aunque requiere recursos lingüísticos avanzados (como modelos de spaCy).

Uso práctico:

* El *stemming* es útil en tareas de búsqueda rápida (por ejemplo, motores de búsqueda).
* La *lematización* es preferible en tareas semánticas (clasificación de texto, análisis de sentimientos, extracción de información).

| Aspecto             | **Stemming**                                         | **Lematización**                                         |
| ------------------- | ---------------------------------------------------- | -------------------------------------------------------- |
| Método              | Reglas de recorte (mecánico)                         | Análisis lingüístico con diccionarios y modelos          |
| Velocidad           | Muy rápida                                           | Más lenta                                                |
| Resultado           | Raíces “recortadas” (a veces no son palabras reales) | Lemas válidos en el idioma                               |
| Precisión semántica | Baja, puede perder significado                       | Alta, conserva significado                               |
| Uso típico          | Motores de búsqueda, indexación rápida               | Clasificación, análisis semántico, comprensión del texto |

Consejos prácticos:

* Usa *stemming* cuando la velocidad sea más importante que la precisión.
* Usa *lematización* cuando necesitas análisis semántico más fino.

## Ejercicio 4 – Morfología (género y número)

Con el texto:

> *“Las alumnas y los profesores estaban contentos.”*

1. Procesa el texto con spaCy y extrae los rasgos de **género** y **número** de cada palabra.
2. Muestra los resultados en una tabla o lista estructurada.
3. Explica cómo esta información podría ser útil en una aplicación de análisis lingüístico (por ejemplo, detección de concordancia gramatical).

**¿QUÉ DEBES HABER ENTENDIDO?**

La **morfología** estudia la forma de las palabras y sus variaciones. En el procesamiento automático del lenguaje, extraer información morfológica permite al sistema conocer rasgos gramaticales como:

* **Género**: masculino, femenino o neutro (cuando aplica).

  * Ej.: *niño* (Masc), *niña* (Fem).
* **Número**: singular o plural.

  * Ej.: *perro* (Sing), *perros* (Plur).

En español, el género y el número se manifiestan principalmente en los sustantivos, adjetivos y determinantes, y son fundamentales para la **concordancia gramatical**.

Ejemplo:

* *“Las alumnas aplicadas”* → todos los elementos (*las, alumnas, aplicadas*) están en **femenino plural**.
* *“El profesor exigente”* → los elementos (*el, profesor, exigente*) están en **masculino singular**.

En NLP, estas marcas morfológicas son útiles para:

* Verificar la **consistencia gramatical** en un texto.
* Construir sistemas de **corrección automática**.
* Enriquecer embeddings con rasgos lingüísticos adicionales.

Librerías como **spaCy** permiten acceder fácilmente a esta información a través del atributo `.morph` de cada token.

**Ejemplo**: género y número en palabras

| Palabra        | Categoría gramatical | Género | Número |
| -------------- | -------------------- | ------ | ------ |
| **niño**       | Sustantivo           | Masc   | Sing   |
| **niña**       | Sustantivo           | Fem    | Sing   |
| **perros**     | Sustantivo           | Masc   | Plur   |
| **profesoras** | Sustantivo           | Fem    | Plur   |
| **el**         | Determinante         | Masc   | Sing   |
| **las**        | Determinante         | Fem    | Plur   |
| **contento**   | Adjetivo             | Masc   | Sing   |
| **contentas**  | Adjetivo             | Fem    | Plur   |

````python
texto_morf = "Las alumnas y los profesores estaban contentos."
````

````python
# NO EJECUTAR HASTA COMPLETAR

# TODO(1): procesa el texto con nlp
doc_morf = nlp(texto_morf)  # nlp(texto_morf)

# TODO(2): extrae token, pos, gender, number
resultado = []
for tok in doc_morf:
    gender = tok.morph.get("Gender")   # tok.morph.get("Gender")
    number = tok.morph.get("Number")   # tok.morph.get("Number")
    resultado.append({
        "token": tok.text,
        "pos": tok.pos_,
        "gender": gender,
        "number": number
    })

for r in resultado:
    print(r)
````

````python
# === CHECKS ===
def find_token(rows, word):
    for r in rows:
        if r["token"].lower() == word:
            return r
    return None

alumnas = find_token(resultado, "alumnas")
profesores = find_token(resultado, "profesores")
assert alumnas and "Fem" in (alumnas["gender"] or []), "'alumnas' debería ser Fem."
assert alumnas and "Plur" in (alumnas["number"] or []), "'alumnas' debería ser Plur."
assert profesores and "Masc" in (profesores["gender"] or []), "'profesores' debería ser Masc."
assert profesores and "Plur" in (profesores["number"] or []), "'profesores' debería ser Plur."
print("✅ Ej.4 OK")
````

## Ejercicio 5 – Reconocimiento de Entidades (NER)

Analiza el siguiente texto con **spaCy**:

> *“Pedro Sánchez se reunió en Madrid con el presidente de Microsoft.”*

1. Identifica las **entidades detectadas** y sus **etiquetas** (PERSON, ORG, LOC/GPE, etc.).
2. Anota el resultado en una lista de pares `(texto, etiqueta)`.
3. Propón un **caso de uso real** donde esta técnica sea útil (por ejemplo, en prensa digital, motores de búsqueda o redes sociales).

**¿QUÉ DEBES HABER ENTENDIDO?**

El **Reconocimiento de Entidades Nombradas** (*Named Entity Recognition, NER*) es una técnica de NLP cuyo objetivo es *detectar y clasificar automáticamente en un texto las menciones a elementos relevantes del mundo real*.

* **Qué son entidades nombradas:**

  * Personas (*Pedro Sánchez*)
  * Lugares (*Madrid*)
  * Organizaciones (*Microsoft*)
  * Fechas, cantidades, productos, etc.

* **Cómo funciona:**
  Herramientas como **spaCy** incluyen modelos preentrenados capaces de segmentar el texto en tokens y asignar etiquetas de entidad (ej: `PERSON`, `GPE`, `ORG`).

* **Ejemplo:**
  Texto: *“Pedro Sánchez se reunió en Madrid con Microsoft”*
  → `Pedro Sánchez` → `PERSON`
  → `Madrid` → `GPE` (*Geopolitical Entity*)
  → `Microsoft` → `ORG`

* **Aplicaciones prácticas:**

  * Minería de datos en prensa y redes sociales.
  * Construcción de bases de conocimiento (ej: Wikipedia, Wikidata).
  * Motores de búsqueda más inteligentes.
  * Sistemas de pregunta-respuesta y chatbots.

El NER es un **puente entre el lenguaje natural y la representación estructurada de datos**, clave para convertir texto libre en información organizada y explotable.

````python
texto_ner = "Pedro Sánchez se reunió en Madrid con el presidente de Microsoft."
````

````python
# NO EJECUTAR HASTA COMPLETAR

# TODO(1): procesa con nlp y extrae entidades
doc_ner = nlp(texto_ner)  # nlp(___)
print("DOC_NER: ", doc_ner.ents)
ents = [(ent.text, ent.label_) for ent in doc_ner.ents]     # [(___.text,___ ent.label_) for ___ in doc_ner.ents]

print("ENTS:", ents)
````

````python
# === CHECKS ===
def normalize(lbl):
    return {"PER": "PERSON", "LOC": "GPE"}.get(lbl, lbl)

labels = {(text, normalize(lbl)) for (text, lbl) in ents}
expect_any = [
    ("Pedro Sánchez", "PERSON"),
    ("Madrid", "GPE"),
    ("Microsoft", "ORG"),
]
hits = sum(int(any((t in s_text and s_lbl == e_lbl) for (s_text, s_lbl) in labels)) for (t, e_lbl) in expect_any)
assert hits >= 2, "Se esperan al menos dos entidades principales correctas."
print("✅ Ej.5 OK")
````
